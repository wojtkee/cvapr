```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np




```

```python
# Za≈Çadowanie danych
dat = pd.read_csv('bigtrain.csv')
data=dat.loc[:,'original_shape_Elongation':]
```

```python
corr_matrix=data.corr()
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(corr_matrix,cmap='coolwarm')
# plt.show()

# plt.savefig('heatmapV1.png' , dpi=1200)




```

```python
target_corr=corr_matrix.loc[:,'target']

top_6_val=target_corr.nlargest(6)
print(top_6_val)
```
